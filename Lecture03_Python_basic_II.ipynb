````python
import math

def get_result_quadratic_equation(a, b, c):
    values = []
    values.append((-b + math.sqrt(b ** 2 - (4 * a * c)) ) / (2 * a))
    values.append((-b - math.sqrt(b ** 2 - (4 * a * c)) ) / (2 * a))
    return values

print(get_result_quadratic_equation(1,-2,1))
````

### if-elif-else문

* 조건이 거짓일 때 연속하여 다른 조건을 검사하는 경우 사용
* if-elif-else문 구조
    ```
    if <condition>:
      <statement1>
      <statement2>
    elif <condition>:
      <statement3>
      <statement4>
    else:
      <statement5>
      <statement6>
    ```

* if-elif-else문을 이용하여 등급 부여
    * score 변수 생성한 뒤 83점 입력
    * 90점 이상이면 "A" 출력

    * 70점 이상 80점 미만이면 "C" 출력
    * 70점 미만이면 "F" 출력

````python
score = 70

if score >= 90 :
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("F")
````

## 매개변수와 리턴

1. 키워드 매개변수
2. 기본 매개변수
3. 가변 매개변수

1. **키워드 매개변수**  
: 함수를 호출할 때 함수 내에 있는 매개변수의 명칭을 직접 지정해서 값을 입력해줄 때 사용하는 매개변수입니다.

* 예시 1 - 매개변수가 1개 있는 경우
    * 함수명: hi
    * 매개변수: n
    * "안녕하세요"를 n회 출력하기


````python
def hi(n):
    for i in range(n):
        print("안녕하세요")
````

````python
hi(n = 3)
````

* 예시 2 - 매개변수가 3개 있는 경우
    * 함수명: pm
    * 매개변수: n1, n2, n3
    * n1 + n2 - n3 계산해서 출력
    

* 예시 3 - 매개변수 3개와 리턴값이 있는 경우
    * 함수명: pm
    * 매개변수: n1, n2, n3
    * n1+n2-n3 계산해서 숫자값 반환

````python
def pm(n1,n2,n3):
    return (n1+n2-n3)
````

````python
pm(1,2,3)
````

2. **기본 매개변수**  
: 매개변수 중에서 `'기본값'`을 갖는 매개변수. `매개변수 = 값`의 형태로, 호출할 때 별도로 입력하지 않아도 됨

* 예시 1 - 매개변수가 1개 있는 경우
    * 함수명: hi
    * 매개변수: n
    * "안녕하세요"를 n회 출력하기

````python
def hi(n=5):
    for i in range(n):
        print("안녕하세요")
````

````python
hi()
````

3. **가변 매개변수**  
* 가변 매개변수는 함수에 하나만 사용가능
* 가변 매개변수 뒤에는 일반 매개변수를 사용할 수 없음

* 예시 - 매개변수 개수를 지정할 수 없는 경우
    * 매개변수 앞에 *를 지정하여 표현 가능
    * 함수명: add
    * 매개변수: *values
    * 받아온 매개변수들을 모두 더함

````python
def add(num, *values):
    total = num
    for value in values:
        total = total + value
    print(total)
````

````python
add(1,2,3)
````

### itertools 모듈

* Python에서 제공하는 자신만의 반복자를 만드는 모듈
* 곱집합, 순열, 조합 등을 산출

함수 | 설명
:--- | :---
`itertools.product(seq1, seq2...)` | 시퀀스의 데카르트 곱(cartesian product)
`itertools.permutations(seq, n)` | 시퀀스에서 n개 원소로 된 순열 생성, n을 지정하지 않거나 n=None으로 하면 최대 길이의 순열 리턴
`itertools.combinations(seq, n)` | 시퀀스의 n개 원소로 된 조합 생성
`itertools.combinations_with_replacement(seq, n)` | 시퀀스의 n개 원소를 중복 허용해 선택하는 조합


````python
import itertools
````

````python
seq1 = [1,2,3]
seq2 = [4,5,6]
for i in itertools.product(seq1,seq2):
    print(i)
````

````python
seq = [1,2,3]
for i in itertools.permutations(seq,2):
    print(i)
````

````python
seq = [1,2,3]
for i in itertools.combinations(seq,2):
    print(i)
````

````python
seq = [1,2,3]
for i in itertools.combinations_with_replacement(seq,2):
    print(i)
````

## 사용자 모듈

* 사용자가 직접 만드는 모듈
* `.py` 형태로 파일 작성
* `%%writefile <파일명>`: 파일 쓰기 가능


* 예시1
    * 모듈명: test1
    * 함수
        * 함수명: add
        * 매개변수: a, b
        * 매개변수 a와 b의 합을 계산

````python
%%writefile test1.py

def add(a, b):
    return a + b
````

* 예시2
    * 모듈명: test2
    * 함수
        * 함수명: mul
        * 매개변수: a, b
        * 매개변수 a와 b의 곱 계산

````python
%%writefile test2.py

def mul(a, b):
    return a * b
````

## 패키지 구성 요소 [참고](https://velog.io/@rhee519/Python-Module-Package)

* `__init__.py`
    * __init__.py 파일은 해당 디렉터리가 패키지의 일부임을 알려주는 역할
    * 패키지 메타데이터 내용 포함
    * 파이썬 3.3 버전부터는 \_\_init\_\_.py 파일 없어도 패키지로 인식
    * 파이썬 3.3 이전버전과의 호환을 위해 \_\_init\_\_.py 파일을 생성
    * \_\_all\_\_ 이라는 리스트 변수에 하위 패키지 이름 작성
    ```
    __all__ = ["sub_pack1", "sub_pack2", "sub_pack3"]
    ```
* `__main__.py`
    * 패키지의 시작점을 알려주는 역할
    * 패키지 자체를 실행하기 위한 용도
    * 패키지를 실행시키면 \_\_main\_\_.py 실행

# 파일 입출력

* 파일 열기 -> 읽기 또는 쓰기 -> 닫기의 순으로 이루어짐
* 파일 열기/닫기
    * `open()` : 파일 생성 후 열기
    * `.close()` : 파일 닫기
* 파일 모드
    * `r` : 읽기 모드 (default)
    * `w` : 쓰기 모드. 기존 파일이 있을 경우 덮어씀
    * `a` : 추가 모드. 기존 파일이 있을 경우 이어서 씀
    * `+` : 읽기/쓰기모드
    * `b` : 이진 파일을 처리

### 텍스트 파일 쓰기

* `write()` : 파일에 쓰기
* `writelines()` : 리스트를 파일에 쓰기
* `!cat <파일명>` : 텍스트 파일 내용 확인


````python
f = open("file.txt","w")
f.write("hello")
f.close()

!cat file.txt
````

````python
a = ["a","b","c"]
f = open("file.txt","w")
f.writelines(a)
f.close()
!cat file.txt
````

````python
a = ["a\n","b\n","c\n"]
f = open("file.txt","w")
f.writelines(a)
f.close()
!cat file.txt
````

* 구구단 3단을 파일에 이어서 쓰기

````python
f = open("file.txt","a")
for i in range(1, 10):
    f.write("3x{0}={1}\n".format(i, 3*i))
f.close()
````

````python
!cat file.txt
````

## 텍스트 파일 읽기

* `read()` : 텍스트 파일을 한꺼번에 읽고 화면에 출력
* `readline()` : 텍스트 파일을 라인 단위로 읽고 화면에 출력
* `readlines()` : 텍스트 파일의 여러 라인을 읽고 화면에 출력. 리스트 형태로 저장



````python
f = open("file.txt","r")
data = f.read()
print(data)
````

````python
f = open("file.txt","r")
f.readline()
````

````python
f = open("file.txt","r")
f.readlines()
````

* `with`문
    * with 블록 내에서 파일을 열고, 벗어나면 자동으로 `close()`를 수행
    

````python
with open("file.txt","r") as f:
    print(f.readline())
````
