## **Colab이란?**
* 클라우드 기반의 무료 jupyter notebook 개발환경
* Colab + Googledrive + Docker + Linux + Googlecloud
* Github와 연동 가능
* 인터넷이 연결된 웹브라우저로 접속 가능

## **노트북 기능**
* 코드셀과 텍스트셀로 분류
* 선택모드와 편집모드로 분류
* 다양한 단축키와 버튼으로 구성
    * `Ctrl + M + H` : 단축키 확인
    * `Ctrl + Shift + P` : 명령 팔레트 확인

* 셀 편집
    * `Enter` : 선택된 셀 편집 가능
    * `Esc` : 셀 편집 종료
    * `Ctrl + M + B` : 현재 셀 하단에 새로운 셀 생성
    * `Ctrl + M + A` : 현재 셀 이전에 새로운 셀 생성
    * `Ctrl + M + D` : 현재 셀/선택항목 삭제
    * `Ctrl + M + M` : 코드셀 -> 텍스트셀로 변환
    * `Ctrl + M + Y` : 텍스트셀 -> 코드셀로 변환
    * `Ctrl + M + -` : 현재 커서에서 셀 분할
    * `Ctrl + M + K` : 현재 셀을 다음 셀 아래로 이동
    * `Ctrl + M + J` : 현재 셀을 이전 셀 위로 이동
    * `Ctrl + Shift + A` : 모든 셀 선택
    

    
* 셀 실행
    * `Shift + Enter` : 셀을 실행하고 다음 셀로 이동, 다음 셀이 없을 경우 새로운 셀 생성
    * `Alt + Enter` : 셀을 실행하고 새로운 셀 생성
    * `Ctrl + Enter` : 셀을 실행하고 해당 셀에 위치
    * `Ctrl + Shift + Enter` : 선택항목만 실행
    * `Ctrl + F9` : 노트의 모든 셀 실행
    * `Ctrl + F10` : 현재 셀 및 그 이후 모든 셀 실행
    * `Ctrl + M + O` : 출력 숨기기 / 숨기기 취소
    * `Ctrl + M + I` : 셀 실행 중단
    * `Ctrl + M + .` : 작업 종료

* 기타
    * `Ctrl + S` : 저장
